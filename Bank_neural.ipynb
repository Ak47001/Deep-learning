```python
df=pd.read_csv("Bank.csv")
```

```python
#preprocessing
x=df.iloc[:,3:13].values
y=df.iloc[:,13].values
```

```python
x.shape
```

```python
y
```

```python
#Encoding the catogorical variable
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_X1=LabelEncoder()
x[:,1]=labelencoder_X1.fit_transform(x[:,1])
x
```

```python
labelencoder_X2=LabelEncoder()
x[:,2]=labelencoder_X1.fit_transform(x[:,2])
x
```

```python
onehotencoder=OneHotEncoder(categorical_features=[1])
x=onehotencoder.fit_transform(x).toarray()
```

```python
x[:,1]
```

```python
x=x[:,1:]
```

```python
x[:,1]
```

```python
#splitiing the dataset into the Training set and test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#feautre scaling important for deep learning
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

```python
#Part 2
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#initializing the ANN
classifier=Sequential()
```

```python
#Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim=6,init='uniform',activation='relu',input_dim=11))
```

```python
#Adding the second hidden layer
classifier.add(Dense(output_dim=6,init='uniform',activation='relu'))
```

```python
#Adding the output layer
classifier.add(Dense(output_dim=1,init='uniform',activation="sigmoid"))
```

```python
#compiling the ANN
#optimizer= adma is function that is give u gradient decent approach and it will reduce the error
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#fitting the ANN to the training set
classifier.fit(X_train,y_train,batch_size=10,nb_epoch=100)
```

```python
#part 3- making the prediction and evaluating the model
#predicting the test set results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
#Making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
